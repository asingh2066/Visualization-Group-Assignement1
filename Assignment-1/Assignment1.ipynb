```python
import plotly.graph_objects as go

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/asingh2066/Visualization-Group-Assignement1/main/Assignment-1/data/NBA-stats.csv')

for col in df.columns:
    df[col] = df[col].astype(str)

df['text'] = df['code'] + '<br>' + \
    'Offensive Efficiency ' + df['OFF_EFF'] + ' Field Goal Percentage ' + df['FG%'] + '<br>' + \
    'Total Points Per State ' + df['PTS'] + ' Rebounds ' + df['REB'] + '<br>' + \
    'Assists ' + df['AST'] + ' Field Goals % ' + df['FG%']

fig = go.Figure(data=go.Choropleth(
    locations=df['code'],
    z=df['AVG_EFF_STATE'].astype(float),
    locationmode='USA-states',
    colorscale='Reds',
    autocolorscale=False,
    text=df['text'], # hover text
    marker_line_color='black', # line markers between states
    colorbar_title="Efficiency of Team By State"
))

fig.update_layout(
    title_text='2019-2020 NBA Teams Per State Breakdown<br>(Hover for breakdown)',
    geo = dict(
        scope='usa',
        projection=go.layout.geo.Projection(type = 'albers usa'),
        showlakes=True, # lakes
        lakecolor='rgb(255, 255, 255)'),
)

fig.show()


#Data for the 2020-2021 Season (ongoing), comparing it to the previous graph to see the effects covid has had on team performance 
#when it comes to statistical values such as points, field goal percentage, effecinecy and etc...
df = pd.read_csv('https://raw.githubusercontent.com/asingh2066/Visualization-Group-Assignement1/main/Assignment-1/data/NBA-stats-covid.csv')
for col in df.columns:
    df[col] = df[col].astype(str)

df['text'] = df['code'] + '<br>' + \
    'Offensive Efficiency ' + df['OFF_EFF'] + ' Field Goal Percentage ' + df['FG%'] + '<br>' + \
    'Total Points Per State ' + df['PTS'] + ' Rebounds ' + df['REB'] + '<br>' + \
    'Assists ' + df['AST'] + ' Field Goals % ' + df['FG%']

fig = go.Figure(data=go.Choropleth(
    locations=df['code'],
    z=df['AVG_EFF_STATE'].astype(float),
    locationmode='USA-states',
    colorscale='Reds',
    autocolorscale=False,
    text=df['text'], # hover text
    marker_line_color='black', # line markers between states
    colorbar_title="Efficiency of Team By State"
))

fig.update_layout(
    title_text='2020-2021 NBA Teams Per State Breakdown<br>(Hover for breakdown)',
    geo = dict(
        scope='usa',
        projection=go.layout.geo.Projection(type = 'albers usa'),
        showlakes=True, # lakes
        lakecolor='rgb(255, 255, 255)'),
)

fig.show()
```
